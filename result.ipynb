```python
import pandas as pd
filename = "result.csv"
df = pd.read_csv(filename)
df = df.rename(columns={'Unnamed: 0': 'Domains', 'avg'='means'})
df = df[['Domains', 'C->A', 'C->W', 'C->D', 
            'A->C', 'A->W', 'A->D', 
            'W->C', 'W->A', 'W->D', 
            'D->C', 'D->A', 'D->W', 'avg']]
df = df.set_index('Domains')
df = df.rename(index={'otda': 'OT', 'uotda': 'UOT', 'uotda_with_cost': 'UOTc'})
df = df.T
df.to_csv(filename)
df
```
